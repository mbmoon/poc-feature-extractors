```python
from PIL import Image

# Відкриваємо BMP-зображення
image = Image.open("Images/sentinel/SS1/Etalon1.bmp")

# Отримуємо всі кольори, що є в зображенні
colors = image.getcolors(maxcolors=256)  # maxcolors визначає, скільки різних кольорів буде знайдено
if colors is None:
    print("Зображення містить більше ніж 256 унікальних кольорів!")
else:
    # Виводимо інформацію про кожен колір
    print("Кольори в зображенні:")
    for count, color in colors:
        print(f"Колір RGB: {color}, кількість пікселів: {count}")
```
